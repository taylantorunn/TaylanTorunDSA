```python
# =========================
# IMPORTS
# =========================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

df = pd.read_csv("flights.csv", nrows=100000)

# =========================
# BASIC CLEANING
# =========================
df = df.dropna(subset=["ARRIVAL_DELAY", "DEPARTURE_DELAY"])

# Create target variable
df["DELAYED"] = (df["ARRIVAL_DELAY"] > 0).astype(int)

# =========================
# BASIC INFO
# =========================
print("\n--- DATA INFO ---")
print(df.info())

print("\n--- DESCRIBE ---")
print(df.describe())


# EDA 1: Airline vs Delay

delay_by_airline = df.groupby("AIRLINE")["DELAYED"].mean()

plt.figure(figsize=(10,5))
delay_by_airline.sort_values().plot(kind="bar")
plt.title("Average Delay Rate by Airline")
plt.xlabel("Airline")
plt.ylabel("Delay Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# EDA 2: Day of Week vs Delay

delay_by_day = df.groupby("DAY_OF_WEEK")["DELAYED"].mean()

plt.figure(figsize=(8,5))
delay_by_day.plot(kind="bar")
plt.title("Delay Rate by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Delay Rate")
plt.tight_layout()
plt.show()


# EDA 3: Departure vs Arrival Delay

plt.figure(figsize=(8,5))
sns.scatterplot(
    x=df["DEPARTURE_DELAY"],
    y=df["ARRIVAL_DELAY"],
    alpha=0.3
)
plt.title("Departure Delay vs Arrival Delay")
plt.xlabel("Departure Delay")
plt.ylabel("Arrival Delay")
plt.tight_layout()
plt.show()


# HYPOTHESIS TEST

print("\n--- HYPOTHESIS TEST: Airline Differences ---")

groups = [group["DELAYED"].values for name, group in df.groupby("AIRLINE")]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)


if p_value < 0.05:
    print("\nResult: Reject H0 -> Airlines have significantly different delay rates.")
else:
    print("\nResult: Fail to reject H0 -> No significant difference between airlines.")
```
